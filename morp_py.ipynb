```python
import nltk
from nltk.stem import PorterStemmer

# Download punkt (optional, for tokenization)
nltk.download('punkt')

# Initialize stemmer
stemmer = PorterStemmer()

# Sample words to analyze morphology (roots + affixes)
words = ["running", "runner", "runs", "easily", "happier"]

print("Word\t\tStemmed (Root form)")

for word in words:
    stem = stemmer.stem(word)
    print(f"{word}\t\t{stem}")
```
