```python
from collections import Counter
text = "I love NLP and I love Python".split()
trigrams = [(text[i], text[i+1], text[i+2]) for i in range(len(text)-2)]
model = {t: c/sum(Counter(trigrams).values()) for t, c in Counter(trigrams).items()}
print(model)
```
